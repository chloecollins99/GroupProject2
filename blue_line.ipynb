```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
```

```python
# Google API Key
from config import gkey
lhp_pd = pd.read_csv("blue_line.csv")
lhp_pd.head()
```

```python
# create a params dict that will be updated with new city each iteration
params = {"key": gkey}
# Loop through the cities_pd and run a lat/long search for each city
for index, row in lhp_pd.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"
    
    address = row['address']
    zip_code = row['zip_code']
    # update address key value
    params['address'] = f"{address},{zip_code}"
    lat_lng = requests.get(base_url, params=params)
#     print(lat_lng.url)
    lat_lng = lat_lng.json()
    lhp_pd.loc[index, "lat"] = lat_lng["results"][0]["geometry"]["location"]["lat"]
    lhp_pd.loc[index, "lon"] = lat_lng["results"][0]["geometry"]["location"]["lng"]
lhp_pd
```

```python
lhp_pd.to_csv("blue_line_coords.csv")
```
